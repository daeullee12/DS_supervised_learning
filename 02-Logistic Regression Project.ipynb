# Logistic Regression Project 

In this project I will be working with a fake advertising data set, indicating whether or not a particular internet user clicked on an Advertisement. I will try to create a model that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

# Logistic Regression

### Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Male']]
y = ad_data['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,
                                                    random_state=101)
```

Training and fitting a logistic regression model on the training set

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

## Predictions and Evaluations

```python
predictions = logmodel.predict(X_test)
```

Creating a classification report for the model.

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

Interpretation:
1. Precision: 
    - For class 0, the precision is 0.91, indicating that 91% of the instances predicted as class 0 correctly classified.
    - For class 1, the precision is 0.94, indicating that 94% of the instances predicted as class 1 correctly classified.
2. Recall:
    - For class 0, the recall is 0.95, indicating that 95% of the actual class 0 instances were correctly classified.
    - For class 1, the recall is 0.90, indicating that 90% of the actual class 1 instances were correctly classified.
3. F1-score:
    - For class 0, the F1-score is 0.93, indicating a balance between precision and recall for class 0.
    - For class 1, the F1-score is 0.92, indicating a balance between precision and recall for class 1.
4. Support:
    - For class 0, the support is 157, meaning there are 157 instances of class 0 in the dataset.
    - For class 1, the support is 143, meaning there are 143 instances of class 1 in the dataset.
5. Accuracy: the model correctly predicted 93% of all instances in the dataset.

Overall, based on this classification report, this logistic regression model demonstrates reasonably high precision, recall and F1-scores for both classes, indicating a good performance in classifying instanes into the respective classes.
